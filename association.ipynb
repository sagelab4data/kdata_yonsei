```python
# Transaction done in United Kingdom
basket_UK = (data[data.Country=='United Kingdom']
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum()
                 .unstack()
                 .fillna(0)
)

# Transaction done in Portugal
basket_Por = (data[data.Country=='Portugal']
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum()
                 .unstack()
                 .fillna(0)
)

# Transaction done in Sweden
basket_Sweden = (data[data.Country=='Sweden']
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum()
                 .unstack()
                 .fillna(0)
)

# Transaction done in Japan
basket_Japan = (data[data.Country=='Japan']
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum()
                 .unstack()
                 .fillna(0)
)
```

```python
# Defining the hot encoding function to make the data suitable 
# for the concerned libraries

def hot_encode (x):
  if (x <=0) :
    return 0
  if (x >= 1) :
    return 1

# Encoding the datasets

basket_France = basket_France.applymap(hot_encode)
```

```python
basket_UK = basket_UK.applymap(hot_encode)
basket_Por = basket_Por.applymap(hot_encode)
basket_Sweden = basket_Sweden.applymap(hot_encode)
basket_Japan = basket_Japan.applymap(hot_encode)
```

### 연습40: 지표 계산

```python
N= len(example)
f_x = sum(['milk' in i for i in example]) # milk
f_y = sum(['bread' in i for i in example]) #bread
f_x_y = sum([all(w in i for w in ['milk','bread'])
             for i in example
            ])

print (
    f" N = {N}\n Freq(X) = {f_x}\n Freq(Y) = {f_y}\n Freq(X,Y) = {f_x_y}"
)
```

```python
N= len(example)
f_x = sum(['milk' in i for i in example]) # milk
f_y = sum(['bread' in i for i in example]) #bread
f_x_y = sum([all(w in i for w in ['milk','bread'])
             for i in example
            ])

print (
    f" N = {N}\n Freq(X) = {f_x}\n Freq(Y) = {f_y}\n Freq(X,Y) = {f_x_y}"
)
```

```python
# f_x_y 행을 직관적으로 변형한 경우,  단, 3개 이상 되는 장바구니를 고려하면 all 사용하는 방식이 유리
N= len(example)
f_x = sum(['milk' in i for i in example]) # milk
f_y = sum(['bread' in i for i in example]) #bread
f_x_y = sum([(('milk' in i) and ('bread'in i))
             for i in example
            ])

print (
    f" N = {N}\n Freq(X) = {f_x}\n Freq(Y) = {f_y}\n Freq(X,Y) = {f_x_y}"
)
```

```python
# support (milk -> bread)
support= f_x_y / N
print (f'support = {support:.4f}')
```

```python
# confidence(milk -> bread)
confidence = support / (f_x / N)
print (f'confidence = {confidence:.4f}')
```

```python
# lift(milk -> bread)
lift = confidence / (f_y/N)
print (f'lift = {lift:.4f}')
```

```python
# leverage(milk->bread)
leverage = support -((f_x/N)*(f_y/N))
print (f'leverage = {leverage:.4f}')
```

```python
# conviction(milk -> bread)
conviction = (1 - (f_y/N)) / (1 - confidence)
print (f'conviction = {conviction:.4f}')
```

```python
def manual_encoding(ll) :
  list_dup_unsort_items = [element for sub in ll for element in sub]
  list_nondup_sort_items = sorted(list(set(list_dup_unsort_items)))

  manual_df = pd.DataFrame(
      False,
      index = range((len(ll))),
      coulumns= list_nondup_sort_items
  )

  for i in range(len(ll)):
    for j in ll[i] :
      manual_df.loc[i,j] = True
  return manual_df
```
